## Divide y vencerás - Problema Torres de Hanói
- Parto de un problema y lo soluciono mediante la solución de subproblemas y unión de resultados
- En este caso concreto: H(n) = H(n-1) + 1 + H(n-1)
- Pensar lo anterior como... "Muevo todos los discos salvo uno a torre auxiliar, muevo el disco más grande a la torre final y muevo todos los discos restantes de la torre auxiliar a la final"
- Solamente resuelvo el problema de mover un disco (condición de parada de división) porque el resto se aplica el mismo algoritmo.
- Mover un disco es muy sencillo -> Solución directa -> Mover de torre de incio a torre de fin
- Ojo, el print del "if" es mover un disco entre torres (cualesquiera) y el print del "else" es el caso de mover el disco grande a torre final

```python
def torres_hannoi(numero_discos, inicio, fin):

    #print(f"torres_hannoi({numero_discos}, {inicio}, {fin})")

    # Si el número de discos es 1 -> Caso base, mover de inicio a fin
    if numero_discos == 1:
        print(f"Caso base: Mover desde {inicio} a {fin} -- torres_hannoi({numero_discos}, {inicio}, {fin})")
        print(f"Mover desde {inicio} a {fin}")
    # Si el número de discos es mayor que 1 -> Mover resto de discos, mover disco único,
    # recuperar => Recursivamente
    else:
        torres_hannoi(numero_discos-1, inicio, 6-inicio-fin)
        print(f"Caso central: Mover desde {inicio} a {fin} -- torres_hannoi({numero_discos}, {inicio}, {fin})")
        print(f"Mover desde {inicio} a {fin}")
        torres_hannoi(numero_discos-1, 6-inicio-fin, fin)

torres_hannoi(3,1,3)
```

## Técnicas voraces - Problema del cambio
- Para resolver el problema, elijo una solución "buena" sin pensar en consecuencias
- Sería como elegir algo "óptimo" localmente con la esperanza de que sea lo mejor en general
- No siempre funciona bien -> Si en el conjunto de las soluciones, dos están próximas entre sí y generan una solución mayor puede fallar (ej: monedas [1, 3, 4, 9])

```python
def cambio_monedas(sistema, cambio):

    print(f"Sistema: {sistema}")
    print(f"Cambio: {cambio}")

    # Inicialización de resultados
    resultado = [0] * len(sistema)

    # Para cada posible moneda, calculamos el número de monedas máximo que podemos dar
    # y lo restamos del cambio pendiente (mejor decisión local, que no tiene por qué
    # ser óptima
    for i in range(len(sistema)):
        total_monedas = cambio//sistema[i] # División con resultado entero
        resultado[i] = total_monedas
        cambio = cambio - resultado[i]*sistema[i]

    return resultado

# Sistema óptmo
sistema = [25, 10, 5, 1]
cambio = 27
resultado = cambio_monedas(sistema,cambio)
print(f"Resultado: {resultado}")

print("")

# Sistema NO óptimo
sistema = [11, 5, 1]
cambio = 15
resultado = cambio_monedas(sistema,cambio)
print(f"Resultado: {resultado}")
```

## Backtracking - Problema de las N reinas
- Se divide el problema en etapas
- En cada etapa se construye una solución parcial
- Se comprueba si esa solución parcial sigue siendo válida
- Si es válida, se continúa con la siguiente etapa
- Si no es válida, se descarta y se prueba otra alternativa
- Si ninguna alternativa funciona, se retrocede a la etapa anterior para cambiar decisiones

```python
# Función que comprueba si por columnas hay un error en la solución
# etapa(i) = etapa (j) -> Error
# También comprueba si por diagonales hay un error en la solución
# (abs(i-j) = abs(xi-xj)) -> Error
def conflicto(solucion, etapa):
    for i in range(etapa+1):
        if solucion.count(solucion[i]) > 1:
            return True
        for j in range(etapa):
            if (i != j):
                if (abs(i-j) == abs(solucion[i]-solucion[j])):
                    return True
    else:
        return False
```

```python
# Búsqueda de una solución
def colocar_reinas(numero_reinas, solucion = None, etapa = 0):

    if solucion is None:
        solucion = numero_reinas*[0]

    # Caso base: si he colocado todas las reinas (he completado todas las etapas)
    # etapa comienza en 0 y llega hasta numero_reinas
    if numero_reinas == etapa:
        print(f"({numero_reinas}, {solucion}, {etapa})")
        print(f"{solucion}")
        return solucion

    else:
        # Recorro todas las columnas posibles para la reina en la fila (etapa)
        for i in range(1,numero_reinas+1):
            solucion[etapa] = i
            #print(f"({numero_reinas}, {solucion}, {etapa})")

            # Compruebo si la colocación actual no entra en conflicto
            # con reinas anteriores (columnas y diagonales)
            if not (conflicto(solucion, etapa)):
                # Si es válida, intento resolver el resto del tablero
                resultado = colocar_reinas(numero_reinas, solucion, etapa + 1)

                # Si desde aquí se encuentra una solución válida,
                # la propago hacia arriba para detener la búsqueda
                # En otras palabras, ya tengo resultado, no necesito
                # seguir probando, lo devuelvo
                if resultado is not None:
                    return resultado

        # Limpio la posición de la reina si todos los estados fallan (vuelta atrás)
        solucion[etapa] = 0

        # Si ninguna columna funciona en esta etapa, esta rama no tiene solución
        #return None

solucion = colocar_reinas(4)
```

```python
def es_prometedora(solucion,etapa):
    for i in range(etapa+1):
        if solucion.count(solucion[i]) > 1:
            return False
        for j in range(i+1,etapa+1):
            if abs(i-j)==abs(solucion[i]-solucion[j]):
                return False
    return True


def reinas(N,solucion=[],etapa=0):
    if len(solucion) == 0:
        solucion = [0 for i in range(N)]
    for i in range(1,N+1):
        solucion[etapa] = i
        if es_prometedora(solucion,etapa):
            if etapa == N-1:
                print(solucion)
            else:
                reinas(N,solucion,etapa+1)
        else:
            None

        solucion[etapa] = 0

reinas(4)
```

Problema: Encontrar los dos puntos más cercanos
- Dado un conjunto de puntos se trata de encontrar los dos puntos más cercanos

Guía para aprendizaje:

- Primer intento: Fuerza bruta
- Calcular la complejidad. ¿Se puede mejorar?
- Segundo intento. Aplicar Divide y Vencerás

```python
# Importacion de librerias necesarias para probar nuestros algoritmos
import random
import math
import time

# Establecemos una semilla para que los numeros aleatorios sean siempre los mismos.
random.seed(42)

# Generacion de datos de prueba 1D, 2D Y 3D usando la libreria random.
# Usamos el metodo random.sample para asegurar que seleccionamos valores unicos y no se repiten.
# --- 1D ---
puntos_1d = random.sample(range(1, 1000000), 100)

# --- 2D ---
puntos_2d_set = set()
while len(puntos_2d_set) < 100:
  punto = (random.randint(1, 10000), random.randint(1, 10000))
  puntos_2d_set.add(punto)
puntos_2d = list(puntos_2d_set)

# --- 3D ---
puntos_3d_set = set()
while len(puntos_3d_set) < 100:
  punto = (random.randint(1, 10000), random.randint(1, 10000), random.randint(1, 10000))
  puntos_3d_set.add(punto)
puntos_3d = list(puntos_3d_set)
```

###***1. PROBLEMA EN 1D*** <br>
<br> *1.1 Fuerza Bruta en 1D*
<br>Consiste en comparar todos los puntos contra todos los puntos para encontrar la menor distancia.

```python
# Creamos una funcion para calcular la distancia entre dos puntos de 1D.
def distancia_1d(p1, p2):
  return abs(p1-p2)

# Esta funcion es el algoritmo de fuerza burta, el cual compara todos los puntos con todos.
def fuerza_bruta_1d(puntos):
  min_dist = float('inf')
  p1, p2 = None, None
  n = len(puntos)

  # Con el primer bucle vamos a tomar un punto 'i'
  for i in range(n):
    #El segundo bucle anidado toma el punto 'j'
    for j in range(i + 1, n):
      # Con el bucle comenzando en 'i + 1', no se compara un punto consigo mismo ni se repite pares.
      d = distancia_1d(puntos[i], puntos[j])
      # Si la distancia es menor a la que se tenia guardada, actualizamos.
      if d < min_dist:
        min_dist = d
        p1, p2 = puntos[i], puntos[j]
  # Retornamos la distancia mas pequena encontrada y los dos puntos involucrados.
  return min_dist, p1, p2

# Finalmente llamamos a la funcion principal e imprimimos el resultado.
print(f"Los puntos de 1D son: {puntos_1d}")
distancia_final, punto_a, punto_b = fuerza_bruta_1d(puntos_1d)
print(f"Fuerza bruta 1D -> Distancia minima: {distancia_final} entre los puntos {punto_a} y {punto_b}")
```

***Calcular la complejidad. ¿Se puede mejorar? <br>***
<br> **Complejidad:** El algoritmo usa dos bucles anidados. El primer bucle se ejecuta *n* veces, y el segundo *n - 1* veces. Al sumar estas iteraciones es una progresion aritmetica que resulta en *O(n^2)*.
<br> **¿Se puede mejorar?**: En lugar de comparar todos con todos, podemos ordenar la lista primero *(O(n log n))* y luego simplemente comparar elementos adyacentes *(O(n))*. El costo total bajaría a *O(n log n)*. O, como nos pide el ejercicio, aplicando Divide y Vencerás.

*1.2 Divide y Vencerás en 1D*
<br> La técnica de divide y vencerás requiere que ordenemos la lista, la dividamos por la mitad, busquemos la menor distancia en la mitad izquierda, en la mitad derecha y, finalmente, comprobemos si hay una pareja de puntos con una distancia menor justo en la "frontera" (un punto de la izquierda y otro de la derecha).

```python
# Funcion recursiva central para Divide y Venceras en 1D
def divide_y_venceras_1d(puntos_ordenados):
  n = len(puntos_ordenados)

  # Si nos quedamos con 3 o menos punto en esta division, es mas rapido usar fuerza bruta para resolver este trozo pequeno.
  if n <= 3:
    return fuerza_bruta_1d(puntos_ordenados)

  # Dividimos la lista en 2 mitades
  mid = n // 2
  izq = puntos_ordenados[:mid]
  der = puntos_ordenados[mid:]

  # Llamamos a esta misma funcion recursivamente para cada mitad.
  dist_izq, p1_izq, p2_izq = divide_y_venceras_1d(izq)
  dist_der, p1_der, p2_der = divide_y_venceras_1d(der)

  # Vemos cual de las dos mitades nos dio la distancia mas pequena.
  if dist_izq < dist_der:
    min_dist, p1, p2 = dist_izq, p1_izq, p2_izq
  else:
    min_dist, p1, p2 = dist_der, p1_der, p2_der

  # Podria existir un par mas cercano cruzando la linea divisoria.
  # Como la lista esta ordenada, solo comparamos el ultimo punto de la izquierda con el primer punto de la derecha.
  dist_frontera = abs(izq[-1] - der[0])
  if dist_frontera < min_dist:
    min_dist, p1, p2 = dist_frontera, izq[-1], der[0]

  return min_dist, p1, p2

# Esta es la funcion principal que prepara los datos antes de aplicar el algoritmo.
def pares_cercanos_1d(puntos):
  # Gracias a ordenar previamente lo cual toma O(n log n), el rendimiento sera mejor.
  puntos_ordenados = sorted(puntos)
  return divide_y_venceras_1d(puntos_ordenados)

# Finalmente llamamos a la funcion principal e imprimimos el resultado.
print(f"Los puntos de 1D son: {puntos_1d}")
dist, p1, p2 = pares_cercanos_1d(puntos_1d)
print(f"Divide y Venceras 1D -> Distancia minima: {dist} entre los puntos {p1} y {p2}")
```

***Calcular la complejidad. ¿Se puede mejorar? <br>***
<br> **Complejidad:** Al aplicar divide y venceras, el paso que domina el tiempo es el ordenamiento inicial de la lista *sorted()*, el cual tiene una complejidad de *O(n log n)*. Las llamadas recursivas dividen el problema a la mitad iterativamente y la combinacion toma tiempo constante *O(1)*. Por lo tanto, la complejidad temporal global del algoritmo baja de *O(n^2)* a *O(n log n)*.
<br> **¿Se puede mejorar?:** *O(n log n)* es el limite inferior optimo para este problema, por lo que no se puede mejorar mas sin usar tecnicas de hashing o caracteristicas particulares del hardware.

***2. PROBLEMA EN 2D <br>***
<br> *2.1 Divide y venceras en 2D*
<br> Para el caso del 2D tenemos que dividir los puntos por la coordenada X. Calculamos la distancia mínima en la izquierda y la derecha. Luego, verificamos los puntos que caen dentro de una "franja central" de ancho igual a la distancia mínima encontrada.

```python
# Calculamos la distancia euclidiana
def distancia_2d(p1,p2):
  return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

# Nuevamente ocupamos fuerza bruta para resolver los casos base o pequenos conjuntos.
def fuerza_bruta_2d(puntos):
  min_dist = float('inf')
  p1, p2 = None, None
  for i in range(len(puntos)):
    for j in range(i + 1, len(puntos)):
      d = distancia_2d(puntos[i], puntos[j])
      if d < min_dist:
        min_dist, p1, p2 = d, puntos[i], puntos[j]
  return min_dist, p1, p2

# Programamos la funcion central recursiva para 2D
def divide_y_venceras_2d(px, py):
  # Si hay 3 puntos o menos
  if len(px) <= 3:
    return fuerza_bruta_2d(px)

  # Encontramos la linea vertical media que divide los puntos en dos mitades.
  mid = len(px) // 2
  punto_medio = px[mid]

  # Separamos en izquierda y derecha basandonos en la coordenada X.
  izq_x, der_x = px[:mid], px[mid:]

  # Tambien separamos los puntos ordenados por Y, manteniendo el orden.
  izq_y = [p for p in py if p[0] <= punto_medio[0]]
  der_y = [p for p in py if p[0] > punto_medio[0]]

  # Buscamos recursivamente en ambas mitades
  d1, p1_1, p2_1 = divide_y_venceras_2d(izq_x, izq_y)
  d2, p1_2, p2_2 = divide_y_venceras_2d(der_x, der_y)

  # Obtenemos la menor distancia de ambas mitades
  d_min, p1, p2 = (d1, p1_1, p2_1) if d1 < d2 else (d2, p1_2, p2_2)

  # Recogemos todos los puntos que estan a una distancia menor a 'd_min' de la linea central.
  # Solo en esta franja podrian existir dos puntos mas cercanos que crucen la linea.
  franja = [p for p in py if abs(p[0] - punto_medio[0]) < d_min]

  # Evaluamos los puntos dentro de la franja.
  for i in range(len(franja)):
    # Por geometria sabemo que un punto en esta franja solo puede estar mas cerca de un maximo de 7
    # vecinos. Esto evita comparar todos contra todos de nuevo.
    for j in range(i + 1, min(i + 8, len(franja))):
      d = distancia_2d(franja[i], franja[j])
      if d < d_min:
        d_min, p1, p2 = d, franja[i], franja[j]

  return d_min, p1, p2

# La ultima funcion nos permitira tener pre-ordenados los datos por X y por Y.
def pares_cercanos_2d(puntos):
  px = sorted(puntos, key=lambda p: p[0])
  py = sorted(puntos, key=lambda p: p[1])
  return divide_y_venceras_2d(px, py)

# Finalmente llamamos a la funcion principal e imprimimos el resultado.
print(f"Los puntos de 2D son: {puntos_2d}")
dist, p1, p2 = pares_cercanos_2d(puntos_2d)
print(f"Divide y Vencerás 2D -> Distancia mínima: {dist:.4f} entre {p1} y {p2}")
```

***3. PROBLEMA EN 3D <br>***
<br> *3.1 Divide y venceras en 3D*
<br> La lógica es una extensión del caso 2D. En lugar de una línea que separa el espacio, usamos un plano. La franja ahora es un "volumen" o "losa". Aunque la geometría es más compleja, la idea de partición es la misma.

```python
# Formula de distancia euclidiana ampliada a 3 ejes (X,Y,Z).
def distancia_3d(p1, p2):
  return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2] - p2[2])**2)

# Fuerza bruta para los casos de 3D.
def fuerza_bruta_3d(puntos):
  min_dist = float('inf')
  p1, p2 = None, None
  for i in range(len(puntos)):
    for j in range(i + 1, len(puntos)):
      d = distancia_3d(puntos[i], puntos[j])
      if d < min_dist:
        min_dist, p1, p2 = d, puntos[i], puntos[j]
  return min_dist, p1, p2

# Funcion recursiva para 3D.
def divide_y_venceras_3d(px, py, pz):
  if len(px) <= 3:
    return fuerza_bruta_3d(px)

  # Encontramos el punto medio en X para usarlo de plano divisor.
  mid = len(px) // 2
  punto_medio = px[mid]

  # Dividimos los puntos en dos mitades a lo largo del eje X.
  izq_x, der_x = px[:mid], px[mid:]

  # Dividimos las listas en Y y Z en base a si estan a la izquierda o derecha del plano central.
  izq_y = [p for p in py if p[0] <= punto_medio[0]]
  der_y = [p for p in py if p[0] > punto_medio[0]]
  izq_z = [p for p in pz if p[0] <= punto_medio[0]]
  der_z = [p for p in pz if p[0] > punto_medio[0]]

  # Llamadas recursivas de la funcion divide y venceras.
  d1, p1_1, p2_1 = divide_y_venceras_3d(izq_x, izq_y, izq_z)
  d2, p1_2, p2_2 = divide_y_venceras_3d(der_x, der_y, der_z)

  # Buscamos el minimo entre ambas mitades.
  d_min, p1, p2 = (d1, p1_1, p2_1) if d1 < d2 else (d2, p1_2, p2_2)

  # Recogemos todos los puntos que estan a una distancia cercana al plano de corte por una distancia menor a d_min.
  franja = [p for p in py if abs(p[0] - punto_medio[0]) < d_min]

  # Evaluamos los puntos dentro de la franja.
  for i in range(len(franja)):
    # Al aumentar una dimension, geometricamente un punto puede tener mas vecinos cercanos.
    # en la franja que en 2D. En teoria, se comprueban los siguientes 15 puntos.
    for j in range(i + 1, min(i + 16, len(franja))):
      d = distancia_3d(franja[i], franja[j])
      if d < d_min:
        d_min, p1, p2 = d, franja[i], franja[j]

  return d_min, p1, p2

# Funcion principal de preparacion 3D.
def pares_cercanos_3d(puntos):
  # Ordenamos el set de punto completo en los 3 ejes distintos.
  px = sorted(puntos, key=lambda p: p[0])
  py = sorted(puntos, key=lambda p: p[1])
  pz = sorted(puntos, key=lambda p: p[2])
  return divide_y_venceras_3d(px, py, pz)

# Finalmente llamamos a la funcion principal e imprimimos el resultado.
print(f"Los puntos de 3D son: {puntos_3d}")
dist, p1, p2 = pares_cercanos_3d(puntos_3d)
print(f"Divide y Vencerás 3D -> Distancia mínima: {dist:.4f} entre {p1} y {p2}")
```
